```python
from iertools.read import read_sql
import pandas as pd
import matplotlib.pyplot as plt
from dateutil.parser import parse
```

```python
f ="../idf/cubo_EPS.sql"
idf = read_sql(f,alias=True)
data = idf.data
data.index = data.index + pd.DateOffset(years=2) + pd.DateOffset(months=-1) + pd.DateOffset(days=-10)

f = "../data/Temixco_EPS_enero_0p3_Este.csv"
eh = pd.read_csv(f,index_col=0,parse_dates=True,usecols=["Time","Ti_1","Tsa_1"])
eh.index = eh.index + pd.DateOffset(hours=-6)  + pd.DateOffset(days=-10)
```

```python
# print(idf.construction_systems)
# idf.get_construction(["EPS"])
```

```python
fig, ax = plt.subplots(figsize=(12,4))

f1 = parse("2025-01-05")
f2 = f1 + pd.Timedelta("1D")

ax.plot(data.Ti_CUBO,label="E+")
ax.plot(data.To,label="Tsa")

ax.plot(eh.Ti_1,label="EH")
ax.plot(eh.Tsa_1,label="Tsa_eh")

ax.set_xlim(f1,f2)
ax.grid()
ax.legend()
```
